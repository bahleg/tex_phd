```python
import matplotlib.pylab as plt
import numpy as np
import scipy.stats as st
%matplotlib inline
```

```python
x = np.arange(0,10)
for l in [1.0, 0.1, 0.01, 0.001]:
    
    y = st.invgamma(l,l).pdf(x)
    plt.plot(x,y, label='$\lambda_1=\lambda_2$='+str(l), lw=3, ls='--')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.rcParams['figure.figsize']=(8,6)
plt.rcParams['font.size']=20
plt.savefig('./invgamma.png')
```
